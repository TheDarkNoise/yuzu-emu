```python
import pandas as pd
from google.colab import data_table
import requests
url = 'https://yuzu-emu.org/game/'
r = requests.get(url)
r.status_code
df = pd.read_html(r.text)
df = df[1]
df = df.drop('Date Tested', 1)
df = df.set_index('Game Title')
```

```python
def f(row):
  if row['Compatibility'] == 'Perfect':
    val = 1
  elif row['Compatibility'] == 'Great':
    val = 2
  elif row['Compatibility'] == 'Okay':
    val = 3
  elif row['Compatibility'] == 'Bad':
    val = 4
  elif row['Compatibility'] == 'Intro/Menu':
    val = 5
  elif row['Compatibility'] == "Won't Boot":
    val = 6
  else:
    val = 7
  return val
```

```python
df['Rating'] = df.apply(f, axis=1)
df = df.sort_values(['Rating', 'Game Title'])
```

```python
df.to_excel('yuzu-comp.xlsx')
```
